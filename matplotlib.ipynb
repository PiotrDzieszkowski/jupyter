```python
salaries = [
    ("Mark", 1000),
    ("John", 1500),
    ("Daniel", 2300),
    ("Greg", 5000)
]
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
X = [0,1,2,3]
Y = [0,1,0,1]
plt.plot(X, Y)
```

```python
salaries
```

```python
names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))
plt.plot(names,salary_values)
```

```python
plt.bar(names,salary_values)
```

```python
plt.title("Salaries")
plt.xlabel("Employee")
plt.ylabel("Salary in zł")
plt.bar(names,salary_values)
```

```python
figure, ax = plt.subplots()
ax.bar(names, salary_values)
ax.set_xlabel("Employee")
ax.set_ylabel("Salary in zł")
```

```python
plt.bar(names,salary_values, color="green")
```

```python
figure, ax = plt.subplots()
ax.barh(names, salary_values,color="green")
```
